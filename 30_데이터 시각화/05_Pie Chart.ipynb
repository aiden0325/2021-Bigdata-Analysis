# 5. Pie Chart
* 카테고리별 또는 그룹별로 집계를 하고 백분율로 표시
* 데이터의 비유을 직관적으로 알 수 있음

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 3, 4]
plt.pie(data)
plt.show()
```

```python
labels = ['순대' ,'치킨', '떡볶이', '초밥']
plt.pie(data, labels=labels)
plt.show()
```

```python
# 퍼센트 출력하기
plt.pie(data, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
explode = [0.2, 0, 0, 0]
colors = ['r', 'g', 'b', 'y']
plt.pie(data, labels=labels, explode=explode)
plt.show()
```
